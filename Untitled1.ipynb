```python
import sys
import PyQt5
from PyQt5 import QtCore, QtGui, uic, QtWidgets
from PyQt5.QtWidgets import *
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Document, Date, Text, connections, Float
```

```python
qtCreatorFile = r"C:\Users\user\Desktop\tax_calc.ui" # Enter file here.

Ui_MainWindow, QtBaseClass = uic.loadUiType(qtCreatorFile)
```

```python
class MyApp(QtWidgets.QMainWindow, Ui_MainWindow):
    def __init__(self):
        QtWidgets.QMainWindow.__init__(self)
        Ui_MainWindow.__init__(self)
        self.setupUi(self)
        self.searchButton.clicked.connect(self.search)
        connections.create_connection(hosts=['localhost'])
        self.client = Elasticsearch()

    def search(self):
        query = self.searchInput.text()
        response = self.client.search(
            index="poem",
            body={
              "query": {
                    "bool": {
                      "must": [{"match": {"poem_text": query}}]            


        }
        },"suggest": {
            "text" : "loverk",
            "my-suggest-1" : {
              "term" : {
                "field" : "poem_title",
                "size":20
              }
        }
        }
        }

        )
        print(query)
        for hit in response['hits']['hits']:
            item = QListWidgetItem(hit['_source']['poem_title'])
            self.listWidget.addItem(item)
   
        
if __name__ == "__main__":
    app = QtWidgets.QApplication(sys.argv)
    window = MyApp()
    window.show()
    sys.exit(app.exec_())
```
