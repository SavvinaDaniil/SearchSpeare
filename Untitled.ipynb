```python
def process(string):
    newstring=''.join(x for x in string if x.isalnum()).lower()
    return newstring
```

```python
def addlist(poemlist,onelist):
    newlist=list(poemlist)
    for i in range(len(poemlist)):
        title1=process(poemlist[i]["title"])
        author1=process(poemlist[i]["author"])
        for j in range(len(onelist)):
            title2=process(onelist[j]["title"])
            author2=process(onelist[j]["author"])
            if title1==title2 and author1==author2:
                print(i,j)
                newlist.remove(poemlist[i])
                break
    return(newlist)
```
