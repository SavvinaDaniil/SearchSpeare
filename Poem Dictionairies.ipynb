POETRY FOUNDATION

```python
directory = os.fsencode(r"C:\Users\user\complete_pf")
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".html"): 
        give_attributes_pf(filename)        
```

```python
def give_attributes_pf(filename):
    try:
        with codecs.open(r"C:\Users\user\complete_pf"+"\\"+filename, 'r', encoding='utf8') as f:
            text = f.read()
    except:
        with codecs.open(r"C:\Users\user\complete_pf"+"\\"+filename, 'r', encoding=None) as f:
            text = f.read()
    soup = BeautifulSoup(text, 'html.parser')
    try:
        result_list = []
        for s in soup.select("div.c-feature-bd"):
            result_list.extend(s.stripped_strings)
        poem=" ".join(result_list).replace(u'\xa0', u' ')
        title=soup.h1.get_text().strip()
        author=soup.a.get_text().strip()
        url=soup.get_text().split('\n')[0]
        dirr={"url":url,"title":title,"author":author,"text":poem}
        poetryfound.append(dirr)
    except:
        print('ax')
```

```python
for i in range(1,len(poetryfound)+1):
    if poetryfound[i-1]['text']=="":
        print(i)
        poetryfound.remove(poetryfound[i-1])
        i=i-1
```

```python
with open('poetryfound.json', 'w') as fp:
    json.dump(poetryfound, fp)
```

POETS ORG

```python
directory = os.fsencode(r"C:\Users\user\complete_poetsorg")
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".html"): 
        give_attributes_poetsorg(filename)  
```

```python
def give_attributes_poetsorg(filename):
    try:
        with codecs.open(r"C:\Users\user\complete_poetsorg"+"\\"+filename, 'r', encoding='utf8') as f:
            text = f.read()
    except:
        with codecs.open(r"C:\Users\user\complete_poetsorg"+"\\"+filename, 'r', encoding=None) as f:
            text = f.read()
    try:
        soup = BeautifulSoup(text, 'html.parser')
        #poem=soup.find("div", {"class":"field field-name-body field-type-text-with-summary field-label-hidden"}).get_text().replace(u'\xa0', u' ')
        #poem=soup.pre.get_text().replace(u'\xa0', u' ')
        result_list = []
        for s in soup.select("pre"):
            result_list.extend(s.stripped_strings)
        poem=" ".join(result_list).replace(u'\xa0', u' ')
        if poem=="":
            result_list = []
            for s in soup.select("p"):
                if s==soup.select("p")[-1]:
                    break
                result_list.extend(s.stripped_strings)
            poem=" ".join(result_list)
            poem=" ".join(result_list).replace(u'\xa0', u' ')
        title=soup.h1.get_text().strip()
        author=soup.span.get_text().strip().split(',')[0]
        url=soup.get_text().split('\n')[0].split('\r')[0]
        dirr={"url":url,"title":title,"author":author,"text":poem}
        poetsorg.append(dirr)
    except:
        print(soup.get_text().split('\n')[0])
    
```

```python
with open('poetsorg.json', 'w') as fp:
    json.dump(poetsorg, fp)
```

FAMOUS POETS

```python
directory = os.fsencode(r"C:\Users\user\complete_fp")
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".html"): 
        give_attributes_fp(filename)  
```

```python
def give_attributes_fp(filename):
    try:
        with codecs.open(r"C:\Users\user\complete_fp"+"\\"+filename, 'r', encoding='utf8') as f:
            text = f.read()
    except:
        with codecs.open(r"C:\Users\user\complete_fp"+"\\"+filename, 'r', encoding=None) as f:
            text = f.read()
    soup = BeautifulSoup(text, 'html.parser')
    try:
        
        poem=text.split('\t')[6].replace(u'\xa0', u' ')
        #poem=soup.pre.get_text().replace(u'\xa0', u' ')
        title=soup.prettify().splitlines()[1]
        author=soup.prettify().splitlines()[2]
        url=soup.prettify().splitlines()[0]
        dirr={"url":url,"title":title,"author":author,"text":poem}
        famouspoets.append(dirr)
    except:
        print(soup.prettify().splitlines()[0])
```

```python
with open('famouspoets.json', 'w') as fp:
    json.dump(famouspoets, fp)
```
